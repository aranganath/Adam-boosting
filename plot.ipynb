```python
import matplotlib.pyplot as plt
%matplotlib inline
import pickle as pkl
batch_size = 512
with open('./results/quasiAdam/'+str(batch_size)+'/Testing_losses/mnist_test', 'rb') as handle:
    a = pkl.load(handle)

with open('./results/adam/'+str(batch_size)+'/Testing_losses/mnist_test', 'rb') as handle:
    b = pkl.load(handle)

a = [a_item.cpu() for a_item in a]
b = [b_item.cpu() for b_item in b]
```

```python
length = 50
plt.plot(a[:length], label='quasiAdam')
plt.plot(b[:length], label='adam')
# plt.plot(d, label='amsgrad')
plt.xlabel('iterations')
plt.ylabel('Prediction accuracy (%)')
plt.legend()
# plt.savefig('./results/figures/batch_size'+str(batch_size)+'MNIST.png')
plt.show()
```
